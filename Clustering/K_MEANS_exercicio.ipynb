Um dos problemas do K-Means é determinar o número de clusters mais adequado. Vamos supor que 2 clusters são suficientes.

Então, criamos dois centroides aleatoriamente para cada um deles e plotamos novamente os pontos, agora com os dois centroides

```python
import math
import random
import statistics as s

# Faixa para temperatura e umidade
temp_min = min(dados["temp"])
temp_max = max(dados["temp"])
umid_min = min(dados["umid"])
umid_max = max(dados["umid"])

# Definicao do numero de centroides
k = 2
c=[(0, 0) for _ in range(k)] # inicializacao
cores_c = ['red', 'orange']


for i in range(k):
  c[i] = (random.uniform(temp_min, temp_max), random.uniform(umid_min, umid_max))

# Plot
plot_dados(c, cores_c)

# Vamos calcular as distâncias de cada um dos pontos até os centroides c1 e c2.
# Em seguida, plotamos novamente os pontos com a cor do centroide do qual estão mais próximos.

# Cálculo da distância quadrática entre dois pontos quaisquer
def calculate_distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Calcula distância de cada ponto para c1 e para c2
def loopable():
  distances_c1 = [calculate_distance(x, y, c[0][0], c[0][1]) for x, y in zip(dados["temp"], dados["umid"])]
  distances_c2 = [calculate_distance(x, y, c[1][0], c[1][1]) for x, y in zip(dados["temp"], dados["umid"])]

  # Associa cada ponto ao cluster mais próximo
  cores_pontos = [cores_c[0] if d1 < d2 else cores_c[1] for d1, d2 in zip(distances_c1, distances_c2)]

  plot_dados(c, cores_c, cores_pontos)

  # Crie uma função para calcular o SSE de cada cluster
  # (soma das distâncias quadráticas de cada ponto do cluster ao centroide).

  # Inicializando os vetores para os clusters
  cluster1 = {'dia': [], 'temp': [], 'umid': []}
  cluster2 = {'dia': [], 'temp': [], 'umid': []}

  # Iterando sobre os dados e comparando as distâncias
  for i in range(len(dados['dia'])):
      if distances_c1[i] < distances_c2[i]:
          # Adicionando ao cluster 1
          cluster1['dia'].append(dados['dia'][i])
          cluster1['temp'].append(dados['temp'][i])
          cluster1['umid'].append(dados['umid'][i])
      else:
          # Adicionando ao cluster 2
          cluster2['dia'].append(dados['dia'][i])
          cluster2['temp'].append(dados['temp'][i])
          cluster2['umid'].append(dados['umid'][i])

  mean_c1 = []
  mean_c2 = []

  mean_c1.append(s.mean(cluster1['temp']))
  mean_c1.append(s.mean(cluster1['umid']))
  mean_c2.append(s.mean(cluster2['temp']))
  mean_c2.append(s.mean(cluster2['umid']))

  print(mean_c1)
  print(mean_c2)

if mean_c1 == c[0] and mean_c2 == c[1]:
  print("clustering at max!")

else:
  c[0] = mean_c1
  c[1] = mean_c2
```
